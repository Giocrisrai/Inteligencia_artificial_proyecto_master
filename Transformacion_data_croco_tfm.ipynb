### Creaci√≥n de DataFrame con la informaci√≥n proveniente del conjunto de datos

```python
s_rho_array = np.array(ds.variables['s_rho'][:])
s_rho_array
```

```python
dt = np.arange(0,ds.variables['time'].size)
dt
```

```python
columns = ['time'+ ' '+ str(time.units),
           'u'+ ' '+ str(u.units),
           'v'+ ' '+ str(v.units),
           'temp'+ ' '+ str(temp.units),
           'salt'+ ' '+ str(salt.units),
           'DIC'+ ' '+ str(DIC.units),
           'TALK'+ ' '+ str(TALK.units),
           'O2'+ ' '+ str(O2.units),
           'CACO3'+ ' '+ str(CACO3.units),
           'PO4'+ ' '+ str(PO4.units),
           'POC'+ ' '+ str(POC.units),
           'Si'+ ' '+ str(Si.units),
           'NANO'+ ' '+ str(NANO.units),
           'ZOO'+ ' '+ str(ZOO.units),
           'DOC'+ ' '+ str(DOC.units),
           'DIA'+ ' '+ str(DIA.units),
           'MESO'+ ' '+ str(MESO.units),
           'BSI'+ ' '+ str(BSI.units),
           'FER'+ ' '+ str(FER.units),
           'BFE'+ ' '+ str(BFE.units),
           'GOC'+ ' '+ str(GOC.units),
           'SFE'+ ' '+ str(SFE.units),
           'DFE'+ ' '+ str(DFE.units),
           'DSI'+ ' '+ str(DSI.units),
           'NFE'+ ' '+ str(NFE.units),
           'NCHL'+ ' '+ str(NCHL.units),
           'DCHL'+ ' '+ str(DCHL.units),
           'NO3'+ ' '+ str(NO3.units),
           'NH4'+ ' '+ str(NH4.units),
           'w'+ ' '+ str(w.units)]
```

```python
def llenar_df(capa):
    
    """Creaci√≥n del DataFrame Vacio"""
    
    data = pd.DataFrame(0, columns = columns,index = dt)
    
    """Llenado del Dataframe con el indice de la capa"""
    for time_index in range(len(data)):
        
        #Data con formato ('time')
        data.loc[time_index,'time second'] = time[time_index]
        
        #Data con formato ('time', 's_rho', 'eta_rho', 'xi_rho')
        data.loc[time_index,'u meter second-1'] = u[time_index,s_rho_array[capa],yj,xj]
        data.loc[time_index,'v meter second-1'] = v[time_index,s_rho_array[capa],yj,xj]
        data.loc[time_index,'temp Celsius'] = temp[time_index,s_rho_array[capa],yj,xj]
        data.loc[time_index,'salt PSU'] = salt[time_index,s_rho_array[capa],yj,xj]
        data.loc[time_index,'DIC umol C L-1'] = DIC[time_index,s_rho_array[capa],yj,xj]
        data.loc[time_index,'TALK umol C L-1'] = TALK[time_index,s_rho_array[capa],yj,xj]
        data.loc[time_index,'O2 umol L-1'] = O2[time_index,s_rho_array[capa],yj,xj]
        data.loc[time_index,'CACO3 umol C L-1'] = CACO3[time_index,s_rho_array[capa],yj,xj]
        data.loc[time_index,'PO4 umol P L-1'] = PO4[time_index,s_rho_array[capa],yj,xj]
        data.loc[time_index,'POC umol C L-1'] = POC[time_index,s_rho_array[capa],yj,xj]
        data.loc[time_index,'Si umol Si L-1'] = Si[time_index,s_rho_array[capa],yj,xj]
        data.loc[time_index,'NANO umol C L-1'] = NANO[time_index,s_rho_array[capa],yj,xj]
        data.loc[time_index,'ZOO umol C L-1'] = ZOO[time_index,s_rho_array[capa],yj,xj]
        data.loc[time_index,'DOC umol C L-1'] = DOC[time_index,s_rho_array[capa],yj,xj]
        data.loc[time_index,'DIA umol C L-1'] = DIA[time_index,s_rho_array[capa],yj,xj]
        data.loc[time_index,'MESO umol C L-1'] = MESO[time_index,s_rho_array[capa],yj,xj]
        data.loc[time_index,'BSI umol Si L-1'] = BSI[time_index,s_rho_array[capa],yj,xj]
        data.loc[time_index,'FER umol Fe L-1'] = FER[time_index,s_rho_array[capa],yj,xj]
        data.loc[time_index,'BFE umol Fe L-1'] = BFE[time_index,s_rho_array[capa],yj,xj]
        data.loc[time_index,'GOC umol C L-1'] = GOC[time_index,s_rho_array[capa],yj,xj]
        data.loc[time_index,'SFE umol Fe L-1'] = SFE[time_index,s_rho_array[capa],yj,xj]
        data.loc[time_index,'DFE umol Fe L-1'] = DFE[time_index,s_rho_array[capa],yj,xj]
        data.loc[time_index,'DSI umol Si L-1'] = DSI[time_index,s_rho_array[capa],yj,xj]
        data.loc[time_index,'NFE umol Fe L-1'] = NFE[time_index,s_rho_array[capa],yj,xj]
        data.loc[time_index,'NCHL mg Chl m-3'] = NCHL[time_index,s_rho_array[capa],yj,xj]
        data.loc[time_index,'DCHL mg Chl m-3'] = DCHL[time_index,s_rho_array[capa],yj,xj]
        data.loc[time_index,'NO3 umol N L-1'] = NO3[time_index,s_rho_array[capa],yj,xj]
        data.loc[time_index,'NH4 umol N L-1'] = NH4[time_index,s_rho_array[capa],yj,xj]
        data.loc[time_index,'w meter second-1'] = w[time_index,s_rho_array[capa],yj,xj]
    
    return data
```

### Guardar la serie temporal en formato CSV

```python
j = s_rho_array.size
for i in range(j):
    df = llenar_df(i)
    df.to_csv('data_croco_febr_marz_capa_'+str(i)+'.csv',index=False)
    print(df)
```
